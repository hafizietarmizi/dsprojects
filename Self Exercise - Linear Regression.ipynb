```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
import seaborn as sns
sns.set()
```

# Data Cleaning

```python
raw_data = pd.read_csv('1.04. Real-life example.csv')
raw_data.head()
```

```python
#to include categorical data in desc stats use include = 'all'
raw_data.describe(include = 'all')
```

```python
#go line by line of desc stats and try to make sense of things, in this example the count is not uniformed, suggesting MV
#unique values of model is too much for regression as it means 312 dummy variables
#there are 3947 yes out of 4345 counts of registration suggesting a potentially weak feature of regression
#based on quantile values, there are potentially outliers due to immense difference between quantile values
```

```python
data = raw_data.drop(['Model'], axis=1)
data.describe(include = 'all')
```

```python
data.isnull().sum()
```

```python
#if you are removing <5% of the observation, you are free to just remove all observations with MV (i.e 172+150/4345*8)
data_no_mv = data.dropna(axis=0)
data_no_mv.describe(include = 'all')
```

```python
sns.distplot(data_1['Price'])
```

```python
sns.distplot(data_no_mv['Mileage'])
```

```python
q = data_1['Mileage'].quantile(0.99)
data_2 = data_1[data_1['Mileage'] <q]
data_2.describe(include ='all')
```

```python
sns.distplot(data_2['EngineV'])
```

```python
data_3 = data_2[data_2['EngineV']<6.5]
data_3.describe(include = 'all')
```

```python
sns.distplot(data_3['EngineV'])
```

```python
sns.distplot(data_3['Year'])
```

```python
q = data_3['Year'].quantile(0.01)
data_4 = data_3[data_3['Year']>q]
sns.distplot(data_4['Year'])
```

# Relaxing Assumptions

## Linearity

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, figsize=(15,3))
ax1.scatter(data_cleaned['Year'], data_cleaned['Price'])
ax1.set_title('Price and Year')
ax2.scatter(data_cleaned['EngineV'], data_cleaned['Price'])
ax2.set_title('Price and EngineV')
ax3.scatter(data_cleaned['Mileage'], data_cleaned['Price'])
ax3.set_title('Price and Mileage')

plt.show()
```

```python
#as the above scatterplots shows exponential relationship instead of a linear, we must transform the data using log method
log_price = np.log(data_cleaned['Price'])
data_cleaned['log_price'] = log_price
data_cleaned
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, figsize=(15,3))
ax1.scatter(data_cleaned['Year'], data_cleaned['log_price'])
ax1.set_title('Log Price and Year')
ax2.scatter(data_cleaned['EngineV'], data_cleaned['log_price'])
ax2.set_title('Log Price and EngineV')
ax3.scatter(data_cleaned['Mileage'], data_cleaned['log_price'])
ax3.set_title('Log Price and Mileage')

plt.show()
```

```python
data_cleaned = data_cleaned.drop(['Price'],axis=1)
```

## No Multicollinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = data_cleaned[['Mileage', 'Year', 'EngineV']]
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif['features'] = variables.columns
vif
```

```python
data_no_multicollinearity = data_cleaned.drop(['Year'], axis=1)
```

## Creating Dummy Variables

```python
data_with_dummies = pd.get_dummies(data_no_multicollinearity, drop_first=True)
data_with_dummies.head()
```

```python
data_with_dummies.columns.values
```

```python
cols = ['log_price', 'Mileage', 'EngineV', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
```

```python
data_preprocessed = data_with_dummies[cols]
data_preprocessed.head()
```

# Linear Regression

```python
targets = data_preprocessed['log_price']
inputs = data_preprocessed.drop(['log_price'], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(inputs)
```

```python
inputs_scaled = scaler.transform(inputs)
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(inputs_scaled, targets, test_size=0.2, random_state=365)
```

```python
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train,y_hat)
plt.xlabel('Target (y_train)', size=18)
plt.ylabel('Prediction (y_hat)', size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
sns.distplot(y_train - y_hat)
plt.title('Residuals PDF', size=18)
```

```python
#based on residual PDF we can see that its almost normally distributed with the only concern of being skewed to the left, 
#this implies the model might overestimate the price but rarely understimate as neg value comes higher predicted values
```

```python
reg.score(x_train,y_train)
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
reg_summary = pd.DataFrame(inputs.columns.values, columns=['Features'])
reg_summary['Weights'] = reg.coef_
reg_summary
```

```python
data_cleaned['Brand'].unique()
```

```python
y_hat_test = reg.predict(x_test)
```

```python
plt.scatter(y_test,y_hat_test, alpha=0.2)
plt.xlabel('Target (y_test)', size=18)
plt.ylabel('Prediction (y_hat_test)', size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
#above plot shows that our model is good at predicting at higher price as it is closer to the 45deg line 1:1 compared to at lower prices
df_pf = pd.DataFrame(np.exp(y_hat_test), columns=['Prediction'])
df_pf.head()
```

```python
df_pf['Target'] = np.exp(y_test)
df_pf.head()
```

```python
y_test = y_test.reset_index(drop=True)
```

```python
df_pf['Target'] = np.exp(y_test)
df_pf
```

```python
df_pf['Residual'] = df_pf['Target'] - df_pf['Prediction']
df_pf['Difference%'] = np.absolute(df_pf['Residual'] / df_pf['Target'] * 100)
df_pf
```
